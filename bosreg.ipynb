```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
dfx=pd.DataFrame(boston.data, columns=boston.feature_names)
dfy=pd.DataFrame(boston.target)

```

```python
dfx.head()
```

```python
dfx
```

```python
dfx.describe()
```

```python
lreg=linear_model.LinearRegression()
```

```python
x_train ,x_test, y_train, y_test=train_test_split(dfx, dfy, test_size=0.3,random_state=42)
```

```python
lreg.fit(x_train,y_train)
```

```python
type(lreg)
```

```python
y_pre=lreg.predict(x_test)
print(y_pre)
```

```python
x_test
```

```python
print(np.mean((y_pre-y_test)**2))
```

```python
plt.scatter(y_test,y_pre,color='green')
plt.plot([0, 50], [0, 50], '--k',color='red')
plt.xlabel('True price ($1000s)',color='red',fontsize='large', fontweight='bold')
plt.ylabel('Predicted price ($1000s)',color='yellow',fontsize='large', fontweight='bold')
```

```python
sns.distplot((y_test-y_pre),bins=36)
```

```python
print("RMS: %r " % np.sqrt(np.mean((y_pre - x_test) ** 2)))
```
